## **Desafio 05: Procurar na documentação o método query(pandas).**

---
**Método Query**: Consulta as colunas de um DataFrame com uma expressão booleana.

```python
dados_filtrados = dados.query('`g-0` > 0')
dados_filtrados.head()
```

## **Desafio 06: Renomear as colunas tirando o hífen.**

```python
dados.columns = dados.columns.str.replace('-', '')
dados.head()
```

## **Desafio 07: Deixar os gráficos bonitões. (Matplotlib.pyplot)**

```python
import matplotlib.pyplot as plt
grafico = dados['tempo'].value_counts().plot.bar()
plt.title('Intervalo de tempo de administração de medicamentos')
plt.xlabel('Intervalos de tempo') #definindo nome do eixo X
plt.ylabel('Total horas')
plt.style.use('ggplot')
plt.rcParams['figure.figsize'] = (10,5)
grafico
```
