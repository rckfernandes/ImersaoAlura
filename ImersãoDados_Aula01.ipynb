## **Desafio 07: Deixar os gr√°ficos bonit√µes. (Matplotlib.pyplot)**

```python
import matplotlib.pyplot as plt
grafico = dados['tempo'].value_counts().plot.bar()
plt.title('Intervalo de tempo de administra√ß√£o de medicamentos')
plt.xlabel('Intervalos de tempo') #definindo nome do eixo X
plt.ylabel('Total horas')
plt.style.use('ggplot')
plt.rcParams['figure.figsize'] = (10,5)
grafico
```
